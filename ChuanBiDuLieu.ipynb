```python
import os
from PIL import Image
import tensorflow as tf
from collections import Counter
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

## Khai báo đường dẫn thư mục


```python
train_dir = "Dataset/train"
val_dir = "Dataset/val"
```

```python
extensions = set()
for root, dirs, files in os.walk(train_dir):
    for file in files:
        ext = os.path.splitext(file)[1].lower()
        extensions.add(ext)
print("Các loại đuôi ảnh trong tập train:", extensions)
```

```python

count_jpg = 0
count_png = 0
for root, dirs, files in os.walk(train_dir):
    for file in files:
        ext = os.path.splitext(file)[1].lower()
        if ext in [".jpg"]:
            count_jpg += 1
        elif ext == ".png":
            count_png += 1
print("Số ảnh JPG:", count_jpg)
print("Số ảnh PNG:", count_png)
```

```python
extensions_val = set()
for root, dirs, files in os.walk(val_dir):
    for file in files:
        ext = os.path.splitext(file)[1].lower()
        extensions_val.add(ext)
print("Các loại đuôi ảnh trong tập val:", extensions_val)
```

```python
count_val_jpg = 0
for root, dirs, files in os.walk(val_dir):
    for file in files:
        ext = os.path.splitext(file)[1].lower()
        if ext in [".jpg"]:
            count_val_jpg += 1
print("Số ảnh JPG trong tập val:", count_val_jpg)
```

```python
mode_counter = Counter()
for root, dirs, files in os.walk(train_dir):
    for file in files:
        path = os.path.join(root, file)
        try:
            with Image.open(path) as img:
                mode = img.mode
                mode_counter[mode] += 1
        except Exception as e:
            print(f"Lỗi khi mở tệp {path}: {e}")

print(" Thống kê số lượng theo mode :")
for mode, count in mode_counter.items():
    print(f"{mode}: {count} ảnh")
```

```python
for root, dirs, files in os.walk(train_dir):
    for file in files:
        path = os.path.join(root, file)

        try:
            with Image.open(path) as img:
                if img.mode != "RGB":
                    rgb_img = img.convert("RGB")
                    rgb_img.save(path)
                    print(f"Đã chuyển: {path} → RGB")
        except Exception as e:
            print(f"Lỗi khi xử lý {path}: {e}")
```

```python
mode_counter_rgb = Counter()

for root, dirs, files in os.walk(train_dir):
    for file in files:
        path = os.path.join(root, file)

        try:
            with Image.open(path) as img:
                mode = img.mode      
                mode_counter_rgb[mode] += 1
        except Exception as e:
            print(f"Lỗi khi xử lý ảnh {path}: {e}")

print(" Thống kê số lượng theo mode:")
for mode, count in mode_counter_rgb.items():
    print(f"{mode}: {count} ảnh")
```

```python
size_counter = Counter()
for root, dirs, files in os.walk(train_dir):
    for file in files:
        path = os.path.join(root, file)
        try:
            with Image.open(path) as img:
                size = img.size  
                size_counter[size] += 1
        except Exception as e:
            print(f"Lỗi khi mở tệp {path}: {e}")
print(" Thống kê số lượng theo kích thước:")
for size, count in size_counter.items():
    print(f"{size}: {count} ảnh")
```

```python
target_size = (224, 224)
for root, dirs, files in os.walk(train_dir):
    for file in files:
        path = os.path.join(root, file)
        try:
            with Image.open(path) as img:
                img = img.resize(target_size)
                img.save(path)
                print(f"Đã chuyển: {path} → Với kích thước {target_size}")
        except Exception as e:
            print(f"Lỗi khi xử lý {path}: {e}")
```

```python
size_counter_clean = Counter()
for root, dirs, files in os.walk(train_dir):
    for file in files:
        path = os.path.join(root, file)
        try:
            with Image.open(path) as img:
                size = img.size  
                size_counter_clean[size] += 1
        except Exception as e:
            print(f"Lỗi khi mở tệp {path}: {e}")
print(" Thống kê số lượng theo kích thước:")
for size, count in size_counter_clean.items():
    print(f"{size}: {count} ảnh")
```
